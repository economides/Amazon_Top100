```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import numpy as np
import datetime

from os import listdir
from os.path import isfile, join


import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

##  -----------  Connect to Webpage and Create Soup  ------------  ##
url = "https://www.amazon.co.uk/Best-Sellers-Electronics-SIM-Free-Mobile-Phones-Smartphones/zgbs/electronics/356496011"
r = requests.get(url)
soup = BeautifulSoup(r.content)
```

Can be run after cell 1

```python
#Get all review tiles
product_tiles = soup.find_all('div', class_ = 'a-section a-spacing-none aok-relative')
```

```python
#How many reviews? - should be 50
len(product_tiles)
```

```python
#Get the required elements
ranking_out = []
titles_out = []
rating_out = []
number_of_revs_out = []
prices_out = []

for product in product_tiles:
    #Get elements
    ranking = product.find("span", {"class": "zg-badge-text"}).text
    titles = product.find("div", {"aria-hidden": "true"}).text.strip()
    prices = product.find("span", {"class": "p13n-sc-price"}).text
    full_box = product.find("span", {"class": "aok-inline-block zg-item"}).text
    
    #Append elements
    ranking_out.append(ranking)
    titles_out.append(titles)
    prices_out.append(prices)
    
#     #Error handling
#     if ranking = None:
#         ranking_out.append('not_found')
#     else:
#         ranking_out.append(ranking)
        
```
